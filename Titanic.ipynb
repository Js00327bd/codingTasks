```python
# replacing null values on the Age column with the relation to the Pclass column
```

```python
def removing_null_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age
       
    
```

```python
df["Age"] = df[["Age", "Pclass"]].apply(removing_null_age, axis= 1)
```

```python
# lets check the heatmap and see if null values on the age column has been replaces or not
sns.heatmap(df.isnull())
```

```python
#on the heatmap we can see there is only one column with null value which is cabin, as i think we do not need the cabin column we will drop it
```

```python
df.drop("Cabin", axis= 1, inplace = True)
```
